## Compare the results generated by all combinations of the threshold 

```python
import os
import numpy
import pandas
```

```python
def compare_files(path_to_file1, path_to_file2, print_messages=False):
    
    # Read File 1
    file_1 = []
    file = open(path_to_file1, "r")
    for item in file.readlines():
        file_1.append(item)
        
    # Read File 2
    file_2 = []
    file = open(path_to_file2, "r")
    for item in file.readlines():
        file_2.append(item)
    
    
    # Check the intersection
    intersection = []
    for item in file_1:
        if item in file_2:
            intersection.append(item)
#     Uncomment if needed
#     print "The number of common messages: " + str(len(intersection))

    # Present in file 1  but not in file 2: 
    # A-B
    list_of_items1 = []
    for item in file_1:
        if item not in file_2:
            list_of_items1.append(item)
#     Uncomment if needed
#     print "The number of messages in file1 but not in file2:" + str(len(list_of_items1))
    
    # Present in file2 but not in file1
    # B-A
    list_of_items2 = []
    for item in file_2:
        if item not in file_1:
            list_of_items2.append(item)
#     Uncomment if needed
#     print "The number of messages in file2 but not in file1:" + str(len(list_of_items2))
    
    if print_messages:
        print "Common messages:"
        print intersection
        
        print "Messsges in file1 and not in file 2"
        print list_of_items1
        
        print "Messages in file2 and not in file 1"
        print list_of_items2
    return len(intersection), len(list_of_items1), len(list_of_items2), len(file_1), len(file_2)
```

```python
dictionary = {}
file_list = []
for file in os.listdir('./output'):
    file_list.append(file)
```

```python
file_list
```

```python
for A in file_list:
    # Store the output in the 'dictionary'. 
    result_dict = {}
    for B in file_list:
        # Return A^B, A-B, B-A
#         if A!=B:
        print "Comparing " + A[:-4] + " with " + B[:-4]
        intersection, a_b, b_a, no_of_entries_in_a, no_of_entries_in_b = compare_files("output/" + A, "output/" + B)

        result_dict['no. of entries'] = no_of_entries_in_a

        # The comparision dictionary
        compare_a_b_dict = {}
        compare_a_b_dict['intersection'] = intersection
        compare_a_b_dict['A-B'] = a_b
        compare_a_b_dict['B-A'] = b_a

        result_dict[B[:-4]] = compare_a_b_dict

        # Update the final dictionary
        dictionary[A[:-4]] = result_dict
```
